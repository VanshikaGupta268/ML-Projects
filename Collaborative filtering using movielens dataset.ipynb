Q6:Find the 5 most similar user for user with user Id 10.


```python
#storing the userId = 10 in a variable
user_10_id=df['userId']==10
df[user_10_id]
```

```python
#Calculating cosine similiarity for userId = 10
every_user=table.values
user_10_id_=every_user[0]
denominator1=np.sqrt(sum([np.square(x) for x in user_10_id]))
cosine_similiarity=[(10,1)]
i=1
for user in every_user[1:]:
    numerator=[x*y for x,y in zip(user_10_id,user)]
    denominator2=np.sqrt(sum([np.square(x) for x in user]))
    costheta = sum(numerator)/(denominator1*denominator2)
    cosine_similiarity.append((table.index[i],costheta))
    i=i+1
cosine_similiarity.sort(key=lambda x:x[1],reverse =True)
similiar_10_users=cosine_similiarity[0:6]
similiar_10_users
```

```python
#mapping the previous results for top 5 similiar users into a table
top_5_users_df = pd.DataFrame()
top_5_users_df
for user in similiar_10_users:
  top_5_users_df = top_5_users_df.append(table.loc[user[0]])
top_5_users_df['costheta'] = [user[1] for user in similiar_10_users]
total_values = top_5_users_df.values
top_5_users_df
```
